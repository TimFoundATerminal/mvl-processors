```python
import matplotlib.pyplot as plt
import numpy as np
```

## Information Density

Why is the information density per digit directly proportional to $log_2$?

```python
# Create an np array of the number bases we want to plot
bases = np.array(range(2, 37))

# Create an np array of the Bits Required for each base digit
bits = np.log2(bases)

# Create a plot of the number of bits required for each base
plt.plot(bases, bits)
plt.xlabel('Base')
plt.ylabel('Bits Per Digit')
plt.title('Information Bits per Digit vs Base')
# show dotted grid on plot
plt.grid(True, which='both', linestyle='--')
# Set zero to be the origin for the y-axis
plt.ylim(bottom=0, top=6)
# Set the origin to be 2 for the x-axis
plt.xlim(left=2)

# Plot a horizontal line at the value of 1.585
plt.axhline(y=1.585, color='r', linestyle='--', label='1.585 bits')

plt.savefig('../figures/bits_per_digit.png')

plt.show()
```
